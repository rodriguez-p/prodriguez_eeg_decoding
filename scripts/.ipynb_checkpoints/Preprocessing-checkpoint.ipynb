# 2. Segmentamos la se√±al en trials (Epochs)

```python
# Cargar los eventos (se guardan como anotaciones en el fichero eeg)
print(set(raw_clean.annotations.description))
```

```python
# Convertir de anotaciones a eventos
events = mne.events_from_annotations(raw_clean, event_id='auto', regexp='Stimulus/')
# print(events)
```

```python
def recode_gabor_events(input_events):
    """
    Esta funci√≥n devuelve una nueva lista de eventos con la informaci√≥n temporal (time_stamps) de los eventos (triggers) de tipo presentaci√≥n de Gabor (c√≥digo=10) que se enviaron durante el experimento,
    pero donde el c√≥digo de los nuevos eventos se reemplaza alguno de los siguientes tipos en funci√≥n del tipo de ensayo:
    - Present/Seen: 55
    - Present/Unseen: 56
    - Absent/Seen: 57
    - Present/Unsee: 59

    Parameters
    ----------
    input_events : numpy array of size (Nx3)
        Array con los eventos del experimento

    Returns
    -------
    Numpy array of size (Nx3)
        Array Numpy con los eventos de tipo presentaci√≥n de Gabor con los c√≥digos recodificados
    """
    good_trial_types = [55, 56, 57, 59]
    event_dict = {'Present/Seen': 55,
                  'Absent/Unseen': 56,
                  'Absent/Seen': 57,
                  'Present/Unseen': 59}
    output_events = []
    # Recorremos todos los eventos de presentaci√≥n del Gabor y vemos el siguiente trigger de anotaci√≥n del trial
    trial_types_array = np.squeeze(input_events[:,2])
    gabor_event_indexs = np.squeeze(np.argwhere(trial_types_array == 10))
    for i in range(0, len(gabor_event_indexs)):
        e = gabor_event_indexs[i]  # √çndice en el vector input_events del trigger actual de presentaci√≥n del Gabor
        next_e = gabor_event_indexs[i+1] if i+1<len(gabor_event_indexs) else None # √çndice en el vector input_events del siguiente trigger de presentaci√≥n del Gabor
        time_stamp = input_events[e, 0]
        # Buscamos entre los triggers que se han mandado en este trial el c√≥digo de alguno de los de good_trial_types
        trial_events = trial_types_array[e+1:next_e]
        mask = np.isin(trial_events, good_trial_types)
        if np.any(mask):
            output_events.append((time_stamp, 0, trial_events[mask][0]))        
#        for i in range(e+1, len(input_events)):
#            etype = input_events[i, 2]
#            if etype==10 or etype==58: #¬†No se ha grabado la respuesta con el tipo de trial o el trial es err√≥neo
#                break
#            elif etype in trial_types:
#                # Creamos un nuevo evento con el tipo de trial
#                output_events.append((time_stamp, 0, etype))
                
    return np.array(output_events), event_dict
```

```python
gabor_events, events_dict = recode_gabor_events(events[0])
gabor_events.shape
```

```python
# Ploteamos con los eventos seleccionamos
raw.plot(events=gabor_events, event_id=events_dict, start=5, duration=10, color='gray', event_color={55:'y', 56:'g', 57:'r', 59:'k'})
```

```python
#¬†Segmentamos la se√±al en √©pochs (ensayos) en base a los eventos
trials = mne.Epochs(raw_clean, gabor_events, event_id=events_dict, tmin=-2.0, tmax=2.0, baseline=(-2.,-0.01), preload=True) 
del raw_clean
trials.save("{}/S{}/{}{:06d}-epo.fif".format(DATA_ROOT , SUBJECT, FILE_PREFIX, SUBJECT), overwrite=True) # guardamos la se√±al segmentada en un archivo
```
